# Attribute Model — Predictions

Load the trained model and see predictions on clothing images.

## 1. Load model

```python
import torch
from attribute_model import create_attribute_model
from attribute_evaluation import predict_image, print_predictions, show_predictions, show_batch_predictions

device = 'mps'
model = create_attribute_model(device)
ckpt = torch.load('attribute_model.pth', map_location=device)
model.load_state_dict(ckpt['model_state_dict'])
model.eval()

print('Model loaded.')
```

## 2. Predict


```python
img = 'data/crops/val/ACDC_Graphic_Tee_img_00000021.jpg'

preds = predict_image(model, img, device)
show_predictions(img, preds)
```
